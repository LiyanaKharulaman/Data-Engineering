```python
def get_info(url):
    r = requests.get(url)
    soup = bs(r.content)
    
    netflix_kv = {}
    
    info_ratings = soup.find_all("div", {"class":"tv-series col-left-center col-full-xs"})
    netflix_kv["title"] = [n.find("h1", {"class":"mop-ratings-wrap__title mop-ratings-wrap__title--top"}).get_text().strip() for n in info_ratings]
    
    tomato = [n.find("div", {"class":"mop-ratings-wrap__half critic-score"}) for n in info_ratings]
    netflix_kv["tomatometer"] = [str(n.find("span", {"class":"mop-ratings-wrap__percentage"}).text.strip().strip("%")) for n in tomato]
    
    audience = [n.find("div", {"class":"mop-ratings-wrap__half audience-score"}) for n in info_ratings]
    netflix_kv["audience_score"] = [str(n.find("span", {"class":"mop-ratings-wrap__percentage"}).text.strip().strip("%")) for n in audience]
    
    return netflix_kv
```

```python
def get_info(u):
    r = requests.get(u)
    r.encoding = 'utf-8'

    html_content = r.text
    soup = bs(html_content, 'html.parser')
    
    netflix_kv = {}

    info_ratings = soup.find_all("div", {"class":"tv-series col-left-center col-full-xs"})
    netflix_kv["title"] = [n.find("h1", {"class":"mop-ratings-wrap__title mop-ratings-wrap__title--top"}).get_text().strip() for n in info_ratings]

    tomato = [n.find("div", {"class":"mop-ratings-wrap__half critic-score"}) for n in info_ratings]
    netflix_kv["tomatometer"] = [str(n.find("span", {"class":"mop-ratings-wrap__percentage"}).text.strip().strip("%")) for n in tomato]

    audience = [n.find("div", {"class":"mop-ratings-wrap__half audience-score"}) for n in info_ratings]
    netflix_kv["audience_score"] = [str(n.find("span", {"class":"mop-ratings-wrap__percentage"}).text.strip().strip("%")) for n in audience]

    return netflix_kv
```

```python
for index, row in enumerate(all_netflix):
    
    if [row["href"] for row in soup.find_all("a", {"class":"article_movie_poster"})]:
        title = [row.find("a").string for row in all_netflix]
        
        relative_path = [row["href"] for row in soup.find_all("a", {"class":"article_movie_poster"})]
        
        bp = ["https:"]
        base_path = bp * len(relative_path)
        
        full_path = ["".join(x) for x in zip(base_path, relative_path)]
        print(full_path)
        print()
        print(len(full_path))
        break
```

```python
for index, u in enumerate(full_path, start = 1):
    try:
        print("Count:",format(index))
        print(u)
        print("done.")
        print()
        netflix_kv_list.append(get_info(u))
            
    except AttributeError as e:
        # find the index that throws an error
        print("Wait! This url throws an error: {}".format(u))
        #print(u.get_text())
        print(e)
        print("The error has been removed.")
        print()

        if index == e: # remove the indexes that have the error
            continue
```
