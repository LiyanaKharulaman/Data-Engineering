### Get Keys and Values

```python
def get_value(row_data):
    # extract "li" in "td"
    if row_data.find("li"):
        return [li.get_text().replace("\xa0", " ") for li in row_data.find_all("li")]
    
    # extract other than "li" in "td" (such as "div")
    else:
        return row_data.get_text().replace("\xa0", " ")
```

```python
# initialize empty string
netflix_kv = {}

for index, row in enumerate(info_box_rows):
    if index == 0:
        # assign title when index = 0
        netflix_kv["title"] = row.find("th").get_text()
        
    elif index == 1:
        continue
    
    else:
        try:
            netflix_key = row.find("th").get_text()
            netflix_value = get_value(row.find("td"))
            netflix_kv[netflix_key] = netflix_value
        
        except AttributeError as e:
            # find the index that throws an error
            print("This index throws an error: {}".format(index))
            print(row.get_text())
            print(e)
            print("The error has been removed.")
            print()
                
            if index == e: # remove the indexes that have the error
                continue
        
print()
print("The final dictionaries:")
print()
print(netflix_kv)
```

#### Get the Infobox

```python
def get_value(row_data):
    # extract "li" in "td"
    if row_data.find("li"):
        return [li.get_text().replace("\xa0", " ").split(" (")[0] for li in row_data.find_all("li")]

    # extract other than "li" in "td" (such as "div")
    else:
        return row_data.get_text().replace("\xa0", " ").split(" (")[0]
```

```python
def cleaning_tag(soup):
    for tag in soup.find_all("sup"):
        tag.decompose()
```

```python
# passing urls into get_info_box function

def get_info_box(url):
    r = requests.get(url)
    soup = bs(r.content)
    
    info_box = soup.find(class_ = "infobox vevent")
    info_box_rows = info_box.find_all("tr")
    
    cleaning_tag(soup)
    netflix_kv = {}
    
    for index, row in enumerate(info_box_rows):
        if index == 0:
            netflix_kv["title"] = row.find("th").get_text()
        
        elif index == 1:
            continue
            
        else:
            try:
                netflix_key = row.find("th").get_text()
                netflix_value = get_value(row.find("td"))
                netflix_kv[netflix_key] = netflix_value
        
            except AttributeError as e:
                # find the index that throws an error
                print("This index throws an error: {}".format(index))
                print(row.get_text())
                print(e)
                print("The error has been removed.")
                print()
                
                if index == e: # remove the indexes that have the error
                    continue
            
    return netflix_kv
```

```python
print(get_info_box("https://en.wikipedia.org/wiki/The_Crown_(TV_series)"))
```

#### Convert to Datetime format for Running Time

```python
print([netflix.get("Running time", "N/A") for netflix in netflix_kv_list])
```

```python
string = "42–7"
newstring = string.split("–")[0]
print(newstring)
```

```python
dash = r"-|–"
mints = r"minutes"
number = r"\d+(-|to|–)?\d+\s({minutes})"

def cleaning_dash(time_string):
    dash_dict = {"-":" ", "–":" "}
    return dash_dict[time_string]

def parse_dash(string):
    time_string = re.search(dash, string)
    replace_dash = cleaning_dash(time_string)
    return replace_dash
```

```python
def minutes_to_integer(running_time):
    if running_time == "N/A":
        return None
    
    #if isinstance(running_time, list):
        #return int(running_time.split(" ")[0])
    
    dash_syntax = re.search(number, running_time)
    
    if dash_syntax:
        return cleaning_dash(dash_syntax.group())
    
    else:
        return None
```

```python
for netflix in netflix_kv_list:
    netflix["Running time (int)"] = minutes_to_integer(netflix.get("Running time", "N/A"))
```
