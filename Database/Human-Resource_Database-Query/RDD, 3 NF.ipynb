#### Retrieving Employee's Biodata.

```python
%sql SELECT * FROM maritalstatus;
```

```python
%%sql
SELECT EmpID,
       printf("%s, %s", First_Name, Last_Name) AS Name,
       printf("%s", Sex) as Gender,
       printf("%s", DOB) as [Date of Birth],
       printf("%s, %s", Zip, State) AS Address,
       printf("%s", ms.maritaldesc) as [Marital Status]
FROM maritalstatus as ms
JOIN HR as h
ON ms.maritalstatusid = h.maritalstatusid
```

#### Retrieving Employee's Department.

```python
%%sql
SELECT
       printf("%s", h.empid) as ID,
       printf("%s, %s", h.First_Name, h.Last_Name) AS Name,
       printf("%s", p.position) as Position,
       printf("%s", d.Department) as Department,
       printf("%s", manager.ManagerName) as [Manager Name],
       printf("%s", i.DateofHire) as [Date of Hiring],
       printf("%s", hired.RecruitmentSource) as [Recruitment Source],
       printf("%s", i.EmploymentStatus) as [Employement Status],
       printf("%s", terminated.DateofTermination) as [Date of Termination],
       printf("%s", terminated.TermReason) as [Termination Reason]
FROM HR as h
JOIN info as i
ON h.empid = i.empid
JOIN post as p
ON p.positionid = i.positionid
JOIN dept as d
ON d.deptid = i.deptid
JOIN manager as manager
ON i.managerid = manager.managerid
JOIN hired as hired
ON h.empid = hired.empid
JOIN terminated as terminated
ON h.empid = terminated.empid
```

#### Company Performance

```python
%%sql
SELECT
      SUM(CASE WHEN performancescore = "Exceeds" THEN 1 ELSE 0 END) as Exceeds,
      SUM(CASE WHEN performancescore = "Fully Meets" THEN 1 ELSE 0 END) as [Fully Meets],
      SUM(CASE WHEN performancescore = "Needs Improvement" THEN 1 ELSE 0 END) as [Needs Improvement],
      SUM(CASE WHEN performancescore = "PIP" THEN 1 ELSE 0 END) as [PIP]
FROM performances as score
JOIN performance as performance
ON score.perfscoreid = performance.perfscoreid;
```

#### Manager Performance

```python
%%sql
SELECT d.department, manager.managername, score.performancescore, COUNT(score.performancescore) as Score,
round(avg(engagementsurvey), 2) as [Average Engagement Survey], round(avg(empsatisfaction), 2) as [Average Employee Satisfaction], specialprojectscount as [Special Project],
dayslatelast30 as [Late Employees], absences as [Absence Employees]
FROM info as i
JOIN dept as d
ON i.deptid = d.deptid
JOIN manager as manager
ON i.managerid = manager.managerid
JOIN performance as performance
ON i.empid = performance.empid
JOIN performances as score
ON performance.perfscoreid = score.perfscoreid
GROUP BY d.department
ORDER BY Score DESC;
```
